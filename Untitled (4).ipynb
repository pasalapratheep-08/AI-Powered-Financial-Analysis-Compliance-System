```python
from crewai import Agent, Task, Crew, Process
from langchain_openai import ChatOpenAI
from pydantic import BaseModel
```

```python
class FinancialReport(BaseModel):
    total_revenue: float
    total_expenses: float
    net_profit: float
    fraud_risk_score: int  # 1-10
    key_insights: list[str]
```

```python
import os
os.environ["OPENAI_API_KEY"]="ENTER_YOUR_API_KEY"
```

```python
financial_analyst = Agent(
    role='Senior Financial Data Analyst',
    goal='Extract and verify transaction patterns from banking data',
    backstory="""You are an expert at spotting trends in messy bank statements. 
    Your precision is unmatched, and you always look for anomalies.""",
    verbose=True,
    allow_delegation=False
)
```

```python
compliance_officer = Agent(
    role='Banking Compliance Officer',
    goal='Ensure all identified patterns meet regulatory standards and flag fraud',
    backstory="""You focus on KYC (Know Your Customer) and AML (Anti-Money Laundering) 
    regulations. You verify the work of the analyst for legal safety.""",
    verbose=True
)
```

```python
analysis_task = Task(
    description='Analyze the last 3 months of transaction data for account {account_id}.',
    expected_output='A breakdown of income, expenses, and potential savings.',
    agent=financial_analyst
)
```

```python
compliance_task = Task(
    description='Review the financial breakdown for any suspicious activity or regulatory breaches.',
    expected_output='A final risk assessment report.',
    output_json=FinancialReport, # Ensures the output is perfectly formatted data
    agent=compliance_officer
)
```

```python
banking_crew = Crew(
    agents=[financial_analyst, compliance_officer],
    tasks=[analysis_task, compliance_task],
    process=Process.sequential # Analyst works first, then Compliance checks it
)
```

```python
result = banking_crew.kickoff(inputs={'account_id': 'ACC-12345'})
print(result)
```
